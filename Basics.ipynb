```python
#In order to find out the count of how many people eat either of the
# three pies for thanksgiving, we make boolean series to see where
# the entries are null.
apple_isnull = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'].isnull()
pumpkin_isnull = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'].isnull()
pecan_isnull = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'].isnull()

#We then combine all of these boolean series into one, where the result is
# false, at least one type of pie was eaten, if true, none of the types
# of pies were eaten.
ate_pies = apple_isnull & pumpkin_isnull & pecan_isnull

print(ate_pies.unique)
```

```python
#We create a function to split each string, if it isn't null by the spaces
# we next take the plus sign off of the 60+ entires, we convert this
# stripped down string into an integer
def convert_age(string):
    if pd.isnull(string):
        return None
    else:
        c1 = string.split(" ")[0]
        c1 = c1.replace("+"," ")
        return int(c1)
        
#We take the function we created and apply it to the Age column to
# loosely estimte the ages of each participant
data['int_age'] = data['Age'].apply(convert_age)
data['int_age'].describe()     

```

```python
#We re work the fucntion in the previous cell to fit our needs to extract
# income data from the column that asks for the combined amount of
# money earned in each participant's hosuehold
def convert_income(string):
    if pd.isnull(string):
        return None
    c1 = string.split(" ")[0]
    if c1 == "Prefer":
        return None
    else:
        c1 = c1.replace('$','')
        c1 = c1.replace(',','')
        return int(c1)
            
data['int_income'] = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(convert_income)          
data['int_income'].describe()       
```
